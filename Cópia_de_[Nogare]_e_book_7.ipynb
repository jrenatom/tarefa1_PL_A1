```r
#install.packages("tydeverse")
library(ggplot2)
```

```r
head(diamonds)
str(diamonds)
```

```r
ggplot(diamonds, aes(x=price, y=carat))
```

Adicionando os pontos com uma geometria

```r
ggplot(diamonds, aes(y=price, x=carat)) +geom_point()
```

Adicionando titulos e legendas


```r
options(repr.plot.width = 7, repr.plot.height = 7)
#adicionando os pontos +geom_point

ggplot(diamonds, aes(x=carat, y=price)) +geom_point() +
 ggtitle("Analise de diamantes", subtitle="Peso x Preço") + 
 xlab("Peso do diamante") +
 ylab("Preço do diamante")
```

Mudando a cor


```r
ggplot(diamonds, aes(x=carat, y=price)) + 
 geom_point(col='darkcyan', size=1) +
 ggtitle("Analise de diamantes", subtitle="Peso x Preço") + 
 xlab("Peso do diamante") +
 ylab("Preço do diamante")
```

Colocando tamanho como variavel

```r
# Como alterar o range de uma variavel continua no aes(size=????)
ggplot(diamonds, aes(x=carat, y=price)) + 
 geom_point(col='darkcyan', aes(size=carat)) +
 ggtitle("Analise de diamantes", subtitle="Peso x Preço") + 
 xlab("Peso do diamante") +
 ylab("Preço do diamante")
```

Adicionando mais uma variavel

```r
ggplot(diamonds, aes(x=carat, y=price)) + 
 geom_point(aes(col=clarity, size=carat)) +
 ggtitle("Analise de diamantes", subtitle="Peso x Preço") + 
 xlab("Peso do diamante") +
 ylab("Preço do diamante")
```

Mudando o tema

```r
ggplot(diamonds, aes(x=carat, y=price)) + 
 geom_point(aes(col=clarity, size=carat)) +
 ggtitle("Analise de diamantes", subtitle="Peso x Preço") + 
 xlab("Peso do diamante") +
 ylab("Preço do diamante") +
 theme_bw()
```

# Visualizações mais avançadas

Linha de tendencia

```r
ggplot(diamonds, aes(x=carat, y=price)) + 
  geom_point() +
  geom_smooth()
```

Na visluação completa


```r
ggplot(diamonds, aes(x=carat, y=price)) + 
 geom_point(aes(col=clarity, size=carat)) +
 ggtitle("Analise de diamantes", subtitle="Peso x Preço") + 
 xlab("Peso do diamante") +
 ylab("Preço do diamante") +
 theme_bw() + 
   geom_smooth(aes(x=carat, y=price, col=clarity))
```

Quebrando em varios graficos

```r
ggplot(diamonds, aes(x=carat, y=price)) + 
 geom_point(aes(col=clarity, size=carat)) +
 geom_smooth(aes(x=carat, y=price, col=clarity)) +
 ggtitle("Analise de diamantes", subtitle="Peso x Preço") + 
 xlab("Peso do diamante") +
 ylab("Preço do diamante") +
 theme_bw() +    
   facet_wrap(~cut)
```

# Cores e grafico de barras

```r
graficoDiamantes <- ggplot(diamonds, aes(x=color, fill=clarity)) + 
  geom_bar(position = 'fill')

graficoDiamantes
```

```r
graficoDiamantes <- ggplot(diamonds, aes(x=color, fill=clarity)) + 
  geom_bar(position = 'fill') +
  xlab("Cores do diamante \n de D (melhor) a J (pior)") +
  ylab("Proporções") +
  labs(fill="Claridade dos diamantes \n de I1 (pior) a IF (melhor)") + 
  ggtitle("Detalhe dos diamantes")

graficoDiamantes
```

Outra Facet
